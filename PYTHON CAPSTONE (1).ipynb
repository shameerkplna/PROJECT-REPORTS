```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv(r'/content/myexcel - myexcel.csv.csv')
df
```

```python
# Replace the existing 'height' column with the new random heights
random_heights = np.random.randint(150, 181, size=len(df))
df['height'] = random_heights
df
```

```python
#Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

team_counts = df['Team'].value_counts()
team_percentages = (team_counts / len(df)) * 100
team_distribution = pd.DataFrame({'Team': team_counts.index, 'Count': team_counts.values, 'Percentage': team_percentages.values})
print(team_distribution)

#Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Percentage', data=team_distribution)
plt.xlabel('Team')
plt.ylabel('Percentage of Employees')
plt.title('Distribution of Employees Across Teams')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Segregate employees based on their positions
position_counts = df['Position'].value_counts()
position_percentages = (position_counts / len(df)) * 100
position_distribution = pd.DataFrame({
    'Position': position_counts.index,
    'Count': position_counts.values,
    'Percentage': position_percentages.values
})
print(position_distribution)

# Bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Position', y='Percentage', data=position_distribution)
plt.xlabel('Position')
plt.ylabel('Percentage of Employees')
plt.title('Distribution of Employees Across Positions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Identify the predominant age group among employees.

if 'Age' in df.columns:
    #Using value_counts()
    age_counts = df['Age'].value_counts()
    predominant_age = age_counts.index[0]
    print(f"The predominant age is: {predominant_age}")

    #Using a histogram
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Age'], kde=True)
    plt.xlabel('Age')
    plt.ylabel('Number of Employees')
    plt.title('Distribution of Employee Ages')
    plt.show()

    #Using age bins
    bins = [0, 20, 30, 40, 50, 60, float('inf')]
    labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60+']
    df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

    age_group_counts = df['Age Group'].value_counts()
    predominant_age_group = age_group_counts.index[0]
    print(f"The predominant age group is: {predominant_age_group}")

    plt.figure(figsize=(10, 6))
    sns.countplot(x='Age Group', data=df)
    plt.xlabel('Age Group')
    plt.ylabel('Number of Employees')
    plt.title('Distribution of Employees by Age Group')
    plt.show()
else:
    print("The 'Age' column is not found in the dataframe.")
```

```python
#Discover which team and position have the highest salary expenditure.

salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

max_expenditure_row = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]

print(f"The team and position with the highest salary expenditure are: {max_expenditure_row['Team']} - {max_expenditure_row['Position']}")
print(f"Total Salary Expenditure: {max_expenditure_row['Salary']}")
```

```python
#Investigate if there's any correlation between age and salary, and represent it visually.

if 'Age' in df.columns and 'Salary' in df.columns:

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Age', y='Salary', data=df)
    plt.xlabel('Age')
    plt.ylabel('Salary')
    plt.title('Correlation between Age and Salary')
    plt.show()

    correlation = df['Age'].corr(df['Salary'])
    print(f"Correlation coefficient between Age and Salary: {correlation}")

else:
    print("Either 'Age' or 'Salary' column is missing in the DataFrame.")
```

```python
#Analyze the distribution of employees across different teams.
plt.figure(figsize=(10, 6))
df['Team'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Employees Across Teams')
plt.ylabel('')
plt.show()

#Analyze the distribution of employees across different positions.
plt.figure(figsize=(12, 6))
sns.countplot(x='Position', data=df)
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.title('Distribution of Employees Across Positions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


#Visualize the distribution of employee ages.
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Employee Ages')
plt.show()


#Identify the team and position with the highest salary expenditure.
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Salary', hue='Position', data=salary_expenditure)
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.title('Salary Expenditure by Team and Position')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#Investigate the correlation between age and salary.
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Salary', data=df)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between Age and Salary')
plt.show()
```
